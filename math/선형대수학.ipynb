# [선형대수학이란?](https://namu.wiki/w/%EC%84%A0%ED%98%95%EB%8C%80%EC%88%98%ED%95%99) 
- 선형대수학은 우리 사회의 여러 문제를 수학적으로 모델링하여 얻은 선형연립방정식을 푸는 수학 이론과 방법들을 연구하는 학문으로, 이를 이용하면 얻어진 해를 원래 사회문제에 대한 답으로 번역하여 문제해결에 도움을 줄 수 있습니다.
- 선형대수학의 연구는 역사적으로 선형적인 컴퓨터를 만드는 데 큰 기여를 했으며, 이 컴퓨터의 발전과 더불어 선형대수학의 연구가 행렬이론을 중심으로 20세기 후반에 가히 폭발적으로 활발해졌습니다. 

# 용어 정리 

## scaler
- 0차원 물리량   
- $scaler \in N(자연수)$
- $scaler \in Z(정수)$
- $scaler \in Q(유리수)$
- $scaler \in R(실수)$

## vector
- 1차원 물리량 / sclaer의 집합(묶음)     
- $a_i \in R^m$: a(벡터)의 i번째 실수 값 
- $vector \in \{1, 2, 3, 4, 5\}$
- $vector \in R^n$

## matrix
- 2차원 물리량 / vector의 집합(묶음)     
- $A_{i,j} \in R^{n,m}$: A(메트릭스)의 i(row)/j(column)의 실수 값
- $matrix \in R^{n,m}$: (n:row, m:column)

# 연산 

## Vector 연산 
a, b를 연산하려면, a와 b의 크기가 같아야 한다.     
($a \in R^n$, $b \in R^n$ )

### 곱셈

- Norm    
벡터의 크기 (Norm) 
  - $ \begin{Vmatrix} a \end{Vmatrix} = \sqrt{\sum{a^2}}$
  - 만약 vector가 [1, 2]라면, 벡터의 크기(Norm)은 $\sqrt{1^2+2^2} = \sqrt{3}$을 나타낸다. 

- 내적     
벡터의 내적은 항상 scaler이다.    
[벡터 간의 유사도(상관계수) 크기](https://www.youtube.com/watch?v=kzqTTCB-Luo)로 해석할 수 있다.     
$cos \theta$가 90도인 경우 내적의 값은 0이며, 이런 관계를 직교(orthogonal)라고 한다.       
a와 b가 직교이면 1차 독립이다. 그러나 a와 b가 1차 독립이라는 것이 꼭 직교를 뜻하는 것은 아니다.
  - $ a \cdot b \in R $
  - $ \begin{Vmatrix} a \end{Vmatrix} \begin{Vmatrix} b \end{Vmatrix} cos \theta $

image.png

- 외적     
c는 a와 b에 대해서 독립이다. 
  - 크기(a x b = c $\in R^m$): $ \begin{Vmatrix} a \end{Vmatrix} \begin{Vmatrix} b \end{Vmatrix} sin \theta $
  - 방향: a와 b에 대해서 직교(orthogonal)이다.


image.png

### 직교 / 1차독립 / 1차종속 

- 직교(orthogonal):     
  - 두 벡터가 서로 직교한다는 것은 두 벡터가 이루는 각이 90도인 경우를 말합니다.
  - 두 벡터의 내적은 0인 경우를 말합니다. 
  - 두 벡터가 직교한다는 뜻은 서로 독립적이라는 뜻입니다.

- 1차 독립(1st order independence):    
  - 두 벡터가 1차 독립이라는 것은 두 벡터가 일직선상에 놓이지 않고 서로 독립적인 방향을 가지고 있다는 것을 말합니다.  
  - 두 벡터가 이루는 각이 0도 또는 180도가 아니라는 뜻입니다.
  - 두 벡터의 내적은 0이 아니며, 내적의 크기는 [유사도(상관관계)](https://www.youtube.com/watch?v=kzqTTCB-Luo)로 해석할 수도 있습니다.

- 1차 종속(1st order dependence):    
  - 두 벡터가 1차 종속이라는 것은 두 벡터가 일직선상에 놓여있다는 뜻입니다.
  - 즉, 하나의 벡터는 다른 벡터의 상수배로 나타낼 수 있습니다. 

image.png

### 수학적 해석 

$ \{a_1, a_2, a_3, ..., a_n \} \in R^n$인 vector들의 모임인 A(Matrix)가 있다면, ($a_i$는 백터임)     

- 1차 독립이라는 뜻은   
> $c_1a_1 + c_2a_2 + ... + c_na_n = 0$인 경우에 $\{ c_1 \sim c_n \}$의 해가 0이 유일하다면, 1차 독립이라고 한다.  

image.png

- 1차 종속이라는 뜻은       
> $c_1a_1 + c_2a_2 + ... + c_na_n = 0$인 경우에 $\{ c_1 \sim c_n \}$의 해가 0이 유일하지 않다면, 1차 종속이라고 한다.

image.png

- [행렬의 계수: rank(A)](https://rfriend.tistory.com/176)라는 뜻은     
> 1차 종속일 때, A($ \{a_1, a_2, a_3, ..., a_n \} \in R^n $)에서 각 vector들의 1차 독립의 수(A)를 말한다.    
> 만약 A에서 vector들 간 1차 독립이 2개 이며, rank(A)=2로 표현할 수 있다.

image.png

image.png

기저는 "$R^m$의 임의의 원소를 표현하기 위해 필요한 최소한의 벡터로 이루어진 집합"입니다.

## [특이값 분해(SVD)](https://angeloyeo.github.io/2019/08/01/SVD.html)

특이값 분해는 임의의 m x n 차원의 행렬 A에 대하여 다음과 같이 행렬을 분해할 수 있다는 '행렬 분해' 방법 중 하나이다.
$$
A = U \Sigma V^T
$$

여기에서 네 행렬 ($A, U, \Sigma V)$는    
- $A$: m x n rectangular matrix
- $U$: m x m orthogonal matrix(직교 행렬)
- $\Sigma$: m x n diagonal matrix(대각 행렬)
- $V$: n x n orthogonal matrix(직교 행렬) 

image.png

## [주성분 분석(PCA)](https://angeloyeo.github.io/2019/07/27/PCA.html)
공분산 행렬은 일종의 행렬로써, 데이터의 구조를 설명해주며, 특히 특징 쌍(feature pairs)들의 변동이 얼마나 닮았는가(다른 말로는 얼마만큼이나 함께 변하는가)를 행렬에 나타내고 있다. 

- 고유 벡터는 그 행렬이 벡터에 작용하는 주축(principal axis)의 방향을 나타내므로 공분산 행렬의 고유 벡터는 데이터가 어떤 방향으로 분산되어 있는지를 나타내준다고 할 수 있다. 
- 고유 값은 고유벡터 방향으로 얼마만큼의 크기로 벡터공간이 늘려지는 지를 예기한다. 따라서 고유 값이 큰 순서대로 고유 벡터를 정렬하면 결과적으로 중요한 순서대로 주성분을 구하는 것이 된다. 

image.png

## [비음수 행렬 분해(NMF)](https://angeloyeo.github.io/2020/10/15/NMF.html)
음수 미포함 행렬 분해(Non-negative Matrix Factorization, NMF)는 음수를 포함하지 않는 행렬 X를 음수를 포함하지 않는 행렬 W와 H의 곱으로 분해하는 알고리즘이다.
$$
X = WH
$$

만약 p개의 feature를 가지고 원래의 데이터셋 X를 분해하고 싶다고 하면,    
- $W \in R^{m,p}$
- $H \in R{p,n}$

이렇게 분해해주게 되었을 때 우선 H부터 생각해보면 H의 각 행은 하나의 feature가 되며, W의 한 행은 각각의 feature들을 얼마만큼 섞어 쓸 것인지에 관한 weight의 의미를 갖게 된다.

image.png

### NMF의 장점 
- 현실 세계의 데이터들 중 상당수는 음의 값이 포함되지 않는다. (예를 들어 그림의 경우 모든 데이터들은 pixel의 세기로 구성되어 있고, 이 값들 중 음수는 없다.)
- 그런데 많은 경우에 사용되는 matrix factorization 방법들(가령, SVD)이나 차원 축소 방법(가량, PCA)에서는 획득할 수 있는 feature들이 음의 값을 갖는 경우가 존재한다. 
- 하지만 NMF는 이들 중에서 유일하게 non-negative feature들을 추출하기 때문에 큰 장점으로 볼 수 있다. 

### PCA와 NMF의 차이 
PCA(또는 SVD)는 feature들 간의 직교성이 보장된다. 즉 전체 데이터의 구조를 설명하는 것에 한계가 있다.
