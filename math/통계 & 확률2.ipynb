## 이항분포(binomial distribution)
이항분포는 확률변수가 배타적인 두 가지 범주를 갖고 각 시행은 독립적인 경우에서의 분포입니다. 이 시행을 `베르누이 시행`이라고도 합니다.

총 n번의 독립적인 베르누이시행에서 관심 있는 범주가 나올 확률이 p라고 한다면,     
이런 조건들이 이항분포를 결정짓는 '상황'이며,      
여기서 이항분포는 관심있는 범주가 나오는 횟수 y를 확률변수로하는 분포입니다.

$$
Y \sim Bin(n,p) 이면,
$$
$$
P(Y=y) = {n \choose y}p^y(1 - p)^{n - y} \qquad y=0, 1, 2, ... , n
$$
$$
{n \choose y} = { n! \over y!(n-y)! } = _{n}\mathrm{C}_{y}(이항계수)
$$

## 포아송분포(poisson distribution)
- 이항분포: n번 중 성공확률이 p일 때, 성공 횟수를 확률변수로 하는 분포 
- 포아송분포: 일정 단위(시간)에서 평균 성공 수가 m(기댓값)일 때, 성공 횟수를 확률변수로 하는 분포 

예를 들어 어떤 공장에서 10시간(일정 단위)마다 평균적으로 2개의 불량품(평균 성공 수 m)이 발생된다면 불량품이 하나도 발생하지 않을 확률부터 수십, 수백개가 발생할 확률까지 성공 횟수에 따른 확률을 다루는 분포입니다.

$$
Y \sim Poisson(m)이면, 
$$
$$
p(Y = y) = { e^{-m}m^y \over y! } \qquad y=0,1,2, ...
$$
- $e$: 자연상수
- $E(Y) = m$: Y에 대한 평균 값 
- $V(Y) = m$: Y에 대한 분산 값 

포아송분포는 특이하게도 평균과 분산이 같습니다. 그래서 빈도 데이터에 적용하기가 적절합니다.

## [정규분포(normal distribution)](https://moni-gram.tistory.com/m/5) 
정규분포는 주어진 데이터의 형태가 산봉우리일 때 가정되는 분포입니다. 흔히 '종모양'에 비유하기도 합니다. 

$$
X \sim N(\mu, \sigma^2)이면,
$$
$$
f(x) = { 1 \over \sigma \sqrt{2\pi} } \exp \left( { -(x - \mu)^2  \over 2\sigma^2 } \right)  \qquad   - \infty < x < \infty
$$
- $E(X) = \mu$: 평균      
데이터의 중심 
- $V(X) = \sigma^2$: 분산($\sigma$: 표준편차)     
데이터의 퍼짐 정도

### [표준정규분포](https://m.blog.naver.com/PostView.naver?isHttpsRedirect=true&blogId=padosori60&logNo=220735271219)
- 표준정규분포에 대한 정의는 모든 정규분포를 평균($\mu$)은 0, 표준편차($\sigma$)는 1이 되도록 표준화한 것이다.
- 어떤 관측값 X의 값이 그 분포의 평균으로부터 표준편차의 몇 배정도나 떨어져 있는 가를 표준화된 정규분포 확률변수 Z로 나타내기 때문에 `Z분포`라고도 한다.
$$
Z = {X - \mu \over \sigma }
$$
  - $Z$: 표준화 된 확률변수 
  - $X$: 표준화 하기 전 확률변수 
  - $\mu$: X의 평균
  - $\sigma$: X의 표준편차

### [중심극한정리](https://bioinformaticsandme.tistory.com/277) 
- 평균이 $\mu$, 표준편차 $\sigma$인 임의의 모집단으로 부터 크기 n인 표본에서의 표본평균은 n이 크면 근사적으로 평균이 $\mu$이고 분산이 ${\sigma^2 \over n}$인 정규분포를 따른다. 
- 모집단이 정규분포라면 표본평균은 표본 n개수와 상관없이 항상 정규분포를 따른다.

## [카이제곱($x^2$) 분포](https://angeloyeo.github.io/2021/12/13/chi_square.html)
양의 정수 k에 대해 k개의 독립적이고 표준정규분포를 따르는 확률변수 $
X_1, X_2, ...., X_k$를 정의하면 자유도 k의 카이제곱 분포는 확률변수의 분포이다.  
$$
Q = \sum{X_i}^2 = X_1^2 + X_2^2 + \cdots + X_k^2 
$$ 

image.png

- 자유도에 따른 카이제곱 분포의 형태    
자유도가 커질 수록 정규분포에 가까워진다. 

image.png

### 카이제곱 분포의 의미 
- 카이제곱 분포는 오차(error) 혹은 편차(deviation)를 분석할 때 도움을 받을 수 있는 분포이다. 
- 중심극한정리에 따라 샘플수가 무수히 많고 합(sum)을 이용해 오차를 정의하는 경우 오차의 분포는 정규분포를 따르게 된다. 
- 그러므로 이렇듯 오차 혹은 편차에 대한 분석을 수행할 때 카이제곱 분포를 이용하면 이 오차 혹은 편차가 우연히 발생할 수 있다고 볼 수 있을만한 수준의 것인지 그렇지 않은지를 판별할 수 있게 되는 것이다. 

### 카이제곱 검정 
> 적합도 검정(goodness-of-fit test)와 교차 분석(cross tabulation analysis)을 이용할 때 카이제곱 분포를 이용하는 경우가 많다.     

피어슨 카이제곱 통계량
$$
x^2 = \sum{ (O_i - E_i)^2 \over E_i }
$$
- $O_i$: i번째 데이터에 대한 관찰값 
- $E_i$: i번째 데이터에 대한 기댓값 
- $(O_i - E_i)^2$: 편차의 제곱 
- ${ (O_i - E_i)^2 \over E_i }$: 특정 정규분포를 따름 

# [통계적 추정](https://m.blog.naver.com/nilsine11202/221382634022)
통계적 추정(Statistical estimation)이란, 모집단에 대한 정보가 없을 때 표본의 성격을 나타내는 통계량을 기초로 하여 모집단의 모수를 추론하는 방법을 뜻한다.

## [표본추출](https://velog.io/@rsj9987/%EA%B0%80%EC%84%A4%EA%B2%80%EC%A6%9D-%EB%B0%8F-t-test)

- 단순 임의 추출(simple random sampling):  
모집단으로부터 샘플을 균등하게 임의로 추출하는 방법 
  - `복원추출`: 추출했던 것을 원래대로 돌려놓고 다시 추출하는 방법(중복 가능)
  - `비복원추출`: 추출했던 것을 원래대로 돌려놓지 않고 다시 추출하는 방법(중복 불가능)

- 체계적 추출(systematic sampling):    
무작위로 배열된 표본에서 시간적 혹은 공간적으로 일정한 간격을 두고 표본을 추출하는 방법 
  - [3, 6, 9, ...] 배열에서 일정한 step을 선택하여 추출 

- 층화 임의 추출(stratified random sampling):   
모집단이 몇 개의 계층으로 구성되어 있을 때 각 계층 원소로부터 임의 추출하는 방법 
  - 성별, 연령대별, 국가별 등 

- 군집 추출(cluster sampling):   
모집단이 여러 군집을 이를 때, 우선 군집을 선택하고 그 집단에서 표본을 추출 
  - 지역 급식 실태조사 -> 지역의 학교 몇 곳을 모집단으로 추출 -> 모집단 학교의 급식 실태 조사 

## 통계적 추정 기준 

- 불편성(Unbiasedness):   
  - 불편성은 비편향성이라고도 하며, 편향(bias)는 치우침을 뜻한다. 
  - 즉 표본으로부터 구한 통계량의 기대치가 추정하려는 모수의 실제 값에 같거나 가까워지는 성질을 의미한다. 

- 효율성(efficiency):   
  - 모수를 추정할 때 여러 개의 불편추정량이 존재한다면, 분산이 가장 작은 추정량을 선택한다는 의미이다. 

- 일치성(consistency):   
  - 표본 크기가 커질 수록 추정량이 모수에 수렴하는 성질을 의미한다.

- 충분성(sufficiency):    
  - 모수에 대해 가능한 많은 표본정보를 내포하는 성질을 의미한다.

### 표준오차(SE; standard error) 
- 표본 통계량의 표준 편차 
- 표본이 매번 추출될 때마다 값이 바뀌는 특성 때문에 표본 통계량은 매번 그 값에 변동(오차)이 있다. 

표준오차는 모집단의 평균 $\mu$, 표준편차 $\sigma$일 때, 다음과 같이 구할 수 있다. 
$$
SE(표준오차) = { \sigma \over \sqrt{n} }
$$
- 표준오차는 표준편차에 비해서 항상 작은 값을 가진다. 
- 표준편차는 모집단의 분포가 얼마나 퍼져있는가에 대한 개념이고, 
- 표준오차는 평균의 추정치에 대한 불확실도를 수치화한 개념이다.

## [신뢰구간](https://m.blog.naver.com/iotsensor/222182891116)
- 유의수준 $\alpha$:    
모수가 신뢰구간 안에 포함되지 않을 확률을 보통 $\alpha$로 표현한다.
- 신뢰수준 $1 - \alpha$:    
모수가 신뢰구간 안에 포함될 확률(신뢰수준)은 $1 - \alpha$로 표현된다. 

$$
P(a < \theta < b) = 1 - \alpha \qquad \theta : 모수 
$$
> 구간 (a,b)을 모수 $\theta$ 에 대한 100(1-$\alpha$) % 신뢰구간이라고 한다. (만약 0.05($\alpha$)라면, 신뢰수준 95%, 신뢰구간 +/- 5%라고 함)

## [가설 검정](https://m.blog.naver.com/PostView.naver?isHttpsRedirect=true&blogId=kwangmo87&logNo=221518004822) 
- 가설 검정은 관찰 대상에 대한 가설을 통계 기법으로 검정하는 것을 의미합니다. 
- 통계적 가설 검정에서는 반드시 귀무가설과 대립가설을 설정해야 합니다. 

`귀무가설(H0)`
> (기존에 알려진)일반적으로 맞다고 가정하는 가설    
> 귀무가설은 차이가 없다, 영향력이 없다, 연관성이 없다, 효과가 없다

`대립가설(H1)`   
> 새롭게 맞다고 증명하려는 가설 
> 대립가설은 차이가 있다, 영향력이 있다, 연관성이 있다, 효과가 있다 

image.png

## 귀무가설의 유의성 검증 
- 귀무가설의 유의성 검증은 데이터를 바탕으로 설정한 귀무가설과 대립가설 중에서 어느 것이 맞는 지를 판단하는 것입니다. 
- 모집단으로부터 표본을 추출하고 통계량으로 귀무가설이 참인지 거짓인지를 판단합니다. 

`1종 오류 $\alpha$(Type 1 Error)`:     
> 1종 오류는 귀무가설이 실제로 참이어서 채택해야 하지만, 표본의 오차로 인해 채택하지 않는 오류입니다.     
> 유의 수준이라고 말하며 $\alpha$(알파)로 표기합니다. 

`2종 오류 $\beta$(Type 2 Error)`:    
> 2종 오류는 귀무가설이 거짓으로 채택하지 말아야 하지만, 표본의 오차로 인해 채택하는 오류입니다.    
> $\beta$로 표기합니다.  

image.png

## 절차 

- 1단계: 귀무가설과 대립가설을 설정합니다.        
- 2단계: 유의 수준 $\alpha$를 정합니다. (0.005 또는 0.01)
- 3단계: 표본을 수집하고 검정 통계량을 계산하고 p값(확률)을 계산합니다. 
- 4단계: p값이 $\alpha$ 보다 작으면 귀무가설을 기각하고, p값이 $\alpha$보다 크면 귀무가설을 채택합니다.
