# Moving average

```python
MA_days = [10, 20, 30]

for ma in MA_days:
    ma_str = "MA: {}".format(ma)
    df_twtr[ma_str] = df_twtr['Adj Close'].rolling(ma).mean()
    df_tsla[ma_str] = df_tsla['Adj Close'].rolling(ma).mean()
    df_msft[ma_str] = df_msft['Adj Close'].rolling(ma).mean()
    df_amzn[ma_str] = df_amzn['Adj Close'].rolling(ma).mean()
```

```python
df_twtr.head()
```

```python
def plot_graphic(company, company_string):
    plt.figure(figsize=(15, 6))
    plt.plot(company['Adj Close'])
    plt.plot(company['MA: 10'])
    plt.plot(company['MA: 20'])
    plt.plot(company['MA: 30'])
    
    plt.title(company_string)
    plt.xlabel('Date')
    plt.ylabel('Price')

    plt.legend(('Adj Close','MA: 10', 'MA: 20', 'MA:30'))
    plt.grid()
    plt.show()
```

```python
data_list = [df_twtr, df_tsla, df_msft, df_amzn]
company_list = ['Twitter', 'Tesla', 'Microsoft', 'Amazon']

for i in range(len(data_list)):   
    plot_graphic(data_list[i], company_list[i])
```
