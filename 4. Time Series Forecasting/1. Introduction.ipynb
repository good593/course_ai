## 횡단면 데이터(Cross sectional data)
> 횡단면 자료(Cross-sectional data) 는 하나의 시점 혹은 기간에 관찰된 여러 변수의 관측치를 의미합니다. 특정 시점 시점이라는 점이 가장 중요한 특징입니다.

위 그림에서는 2000년의 각 기업 매출액이 바로 횡단면 자료입니다.

## 종단면 데이터(시계열 데이터; Time-series data)
> 시계열 자료(종단면 자료, Time-series data)는 하나의 변수를 여러 시점에 대해 관측한 자료를 의미합니다. 횡단면 자료와 다른 점은 하나의 변수에 대해 여러 시점 혹은 기간 동안 관측된 자료라는 점입니다.

위 그림에서는 기업 A의 연도별 매출액이 바로 시계열 자료인 것입니다.

## 패널 데이터(Panel data)
> 패널 자료(Panel data) 는 횡단면 자료와 시계열 자료가 혼합된 형태라고 이해하면 됩니다. 여러 시점에 대해 관측된 여러 변수의 자료가 모두 존재하는 경우입니다.

위 그림에서 보이는 표, 즉 기업 A, B, C, D 의 2000년 부터 2004년 까지 관측된 모든 매출액 자료 전체가 바로 패널 자료인 것입니다.

# [정상성(stationarity)](https://assaeunji.github.io/statistics/2021-08-08-stationarity/)
- 시계열 자료를 분석하는 통계적인 방법에서 빠질 수 없는 개념은 정상성 (Stationarity)입니다. 
- 그 이유는 정상성은 시계열 분석을 할 때 필수적으로 고려하는 가정이기 때문이죠. 
- 마치 회귀 분석을 할 때 오차의 정규성, 등분산성을 만족해야 그 회귀 분석 결과를 믿을 수 있는 것처럼, `시계열 분석을 할 때에도 정상성을 만족해야 분석 결과에 대해 신뢰할 수 있습니다.`

## 정상성의 정의 
- 정상성 (定常性)은 “안정되다”라는 뜻을 가진 정 (定)과 “항상”이라는 뜻을 가진 상 (常)의 한자를 가져, “일정하여 늘 한곁같은 성질“을 의미합니다. 
- 이 단어 뜻에 맞게 정상성을 가진 시계열은 과거와 현재와 미래의 모두 항상 안정적인, 일정한 분포를 가진 것이 특징입니다. 

image.png

여기서 정상성을 가지는 시계열은 (b)와 (g)뿐입니다.
- (d),(h)는 계절성이 보이기 때문에 정상성을 띠지 않습니다.
- (a),(c),(e),(f)는 어떤 특정한 트렌드 (추세)가 있기 때문에 정상성을 띠지 않습니다.
- (i)는 계절성도, 트렌드도 보이며 심지어 분산도 시간에 따라 커지는 형태이므로 정상성을 띠지 않습니다.

결과적으로 정상성은 어떤 특정한 주기로 반복하는 계절성이나 위로 혹은 아래로 가는 트렌드가 없어야 하며, 분산도 커지면 안되는 것이라 유추 할 수 있습니다.     
즉, 시간에 무관하게 과거, 현재, 미래의 분포가 같아야 한다는 의미입니다.

## 수학적의 정상성 정의 

> 정상성이란 시계열의 평균과 분산이 일정하고, 특정한 트렌드 (추세)가 존재하지 않는 성질을 의미합니다.

정상성은 강한 정상성 (Strong Stationarity)와 약한 정상성 (Weak Stationarity)로 정의가 되는데, 일반적으로 약한 정상성만 만족해도 분석이 타당하다고 보기 때문에, 약한 정상성에 대해서만 더 자세히 알아보겠습니다.

약한 정상성을 수학적으로 더 엄밀하게 정의한다면 다음의 세 가지 조건을 만족해야 합니다.
1. 임의의 $t$에 대해서 $E(X_t) = \mu$
2. 임의의 $t$에 대해서 $Var(X_t) < \infty$
3. 임의의 $t$, $h$에 대하여 $Cov(X_{t+h}, X_t) = \gamma(h)$

3번에서 $\gamma(h)$는 자기공분산 함수 (Autocovariance Function, ACVF)라 불리우는데, 공분산이 시점 $t$에 의존하지 않고, 시간의 차이인 $h$에만 의존함을 의미합니다.

예를 들어, A 주식이 정상성을 띤다고 가정합시다. 이 주가의 평균이 3천원이고, 임의의 시점을 $t$ = 2020년 1월 3일이라 치면
- 2020년 1월 3일의 주가의 평균이 $\mu = 3천 원$이고
- 2020년 1월 3일의 주가의 분산이 유한하며 (= 극단치로 터지지 않고)
- 2020년 1월 3일의 주가와, 그로부터 5일 뒤인 2020년 1월 8일의 주가 간의 공분산이 $\gamma(5)$임을 의미합니다.


## 정상성이 중요한 이유 
- 정상성이 중요한 이유는 시계열의 평균과 분산이 일정해야 시계열 값을 예측할 수 있기 때문입니다. 
- 시계열 분석에서는 “우리가 구한 시계열 자료”가 어떤 시간에 따른 확률 과정 (stochastic process)에서 실현된 값이라고 보는데요.

### 확률 과정(stocastic process)
- 확률 과정은 시간의 흐름에 따라 확률 변수가 변화하는 것을 모델링하는 수학적 도구입니다. 즉, 확률 변수의 집합이 시간에 따라 변화하는 것을 의미합니다.
- 확률 과정은 이산적(discrete)인 경우와 연속적(continuous)인 경우가 있습니다. 이산적인 경우에는 시간이 이산적인 값으로 정의되며, 연속적인 경우에는 시간이 연속적인 값으로 정의됩니다.

# 시계열 데이터 분석 목적
일반적으로 시계열 데이터 분석을 통해서 이상치 탐지 또는 데이터 예측을 하는 목적으로 사용을 하고 있습니다.


## [이상치 탐지(Anomaly Detection)](https://velog.io/@vvakki_/Anomaly-Detection%EC%9D%B4%EC%83%81%EC%B9%98-%ED%83%90%EC%A7%80%EB%9E%80)


시계열 데이터에서 비정상이란 것은 예상과 다른 데이터를 의미합니다. 그리고 그런 비정상 데이터는 크게 3가지로 분류할 수 있습니다. 
1. 평소보다 데이터가 심하게 크거나 작은 경우
2. 일시적인 데이터의 패턴 변경
3. 데이터의 크기 변경 

image.png

### 이상치 탐지 적용 분야 
- Fraud Detection System(이상 거래 탐지):   
신용, 카드, 보험과 같은 금융계에서 불법 및 악용 사례를 탐지 
- Intrusion Detection System(침입 탐지 시스템):    
컴퓨터 또는 네트워크 시스템에 대한 원치 않는 조작을 탐지  
- Predictive Maintenance(설비 예지 보전):    
설비가 고장나기 전, 이상 신호를 탐지  
- Customer Churn(고객 이탈):   
제품 또는 서비스에서 구매와 사용을 중지하는 고객을 분석하는 분야 

### 이상치 탐지 종류 
- Supervised Anomaly Detection:       
학습 데이터에 비정상 데이터와 정상 데이터의 label이 모두 존재했을 때의 경우를 말합니다.    
- Semi-supervised Anomaly Detection:     
Semi-supervised Anomaly Detection은 정상 데이터만을 가지고 학습한 경우를 말합니다. 즉 현재 갖고 있는 데이터가 모두 정상 데이터이지만, 미래의 이상 징후를 탐지하고 싶은 경우에 적절합니다.
- Unsupervised Anomaly Detection:      
학습 데이터에 정상, 비정상 데이터의 labeling이 안된 경우를 말합니다. 즉, 데이터에 대한 가정이 없을 경우로서 모든 데이터를 활용하여 학습을 하긴 하지만, 모델링 후 정상/비정상에 대한 구분이 필요할 수 있습니다. 

### [이상치 탐지 방법론](https://blog.ex-em.com/1146)
데이터 분석은 같은 데이터의 모양이라고 하더라도 도메인에 따라서 접근법이 많이 다룰 수 있습니다. 결국 고객이 무엇을 원하는지 요구사항을 잘 파악하여야 문제를 잘 정의하고 이에 따른 분석방법 및 해결책이 나올 수 있습니다.  

- STL decomposition    
STL decomposition은 시계열 데이터를 Sessional, Trend and Residual로 분해하여 분석하는 알고리즘입니다.

STL은 트랜드를 찾아내는 곳에서도 사용될 수 있지만, Residual Graph를 잘 보면 비정상포인트를 찾을 수 있습니다. 

<장점>    
장기적 데이터에서 뚜렷한 주기, 트랜드를 구분 짓고 구현이 쉽다.    
<단점>     
데이터가 많이 출렁이거나 등락이 강할 경우에 트랜드함을 가지지 못해 분석 결과를 결론 짓기 애매한 경우가 많다.

정상과 비정상 데이터가 레이블링이 되어있는 데이터를 가지고 있을 때 사용할 수 있습니다. CART의 종류로는 xgboost, catboost 등이 있습니다.

<장점>    
Supervised learning이므로 다른 알고리즘보다 한단계 더 직관적결과를 얻을 수 있습니다.  
<단점>     
labeling data가 없으면 분석이 불가능합니다.

- Moving Average(이동평균선)      
Moving Average(이동평균선)은 데이터의 추세를 볼 수 있는 가장 고전적이고 쉬운 방법입니다. 
  - 일정기간 동안의 데이터 평균값을 이동평균선이라고 정의하며, 
  - 이때 각 데이터별 표준편차값을 이용하여 신뢰구간을 그린 다음에 
  - 실제 데이터들이 이 신뢰구간을 벗어난 유무를 이용하여 이상치를 탐지합니다.

<장점>     
계산이 빠르고 직관적이고 어느 데이터에서든 사용이 가능합니다.  
< 단점>   
많은 데이터가 필요하고 사용자의 경험치(윈도우 사이즈 결정)가 중요합니다.

<장점>

<단점>

  - 새로운 ㄷ이터를 분석하여 기존에 가지고 있던 Clustering(군집)에 이상치를 벗어난 데이터들을 발견해낼 수 있습니다.    

<장점>

<단점>

- [LSTM AE](https://velog.io/@jonghne/LSTM-AE%EB%A5%BC-%EC%9D%B4%EC%9A%A9%ED%95%9C-%EC%8B%9C%EA%B3%84%EC%97%B4-%EB%8D%B0%EC%9D%B4%ED%84%B0-%EC%9D%B4%EC%83%81-%ED%83%90%EC%A7%80-1-%EA%B0%9C%EC%9A%94)

3. 재구성 오류를 사전에 정의한 Threshold(임계값)과 비교하여 임계값 초과 시 이상 데이터라고 판단한다.

## 데이터 예측 

# [Forecasting(예측)](https://otexts.com/fppkr/intro.html)

예측은 수천년전부터 지금까지 사람들의 마음을 사로잡고 있습니다. 어떤 때는 예측이 신적인 영감의 흔적으로 간주되기도 하고, 어떤 때는 범죄 행동으로 여겨지기도 했습니다. 유대인 선지자 이사야는 기원전 700년 경에 이렇게 적었습니다.

```text
뒤에 올 일을 알게 하라 그리하면 너희가 신들인 줄 우리가 알리라 
또 복을 내리든지 재난을 내리든지 하라 우리가 함께 보고 놀라리라
(이사야 41:23; 개역개정판)
```

컴퓨터에 관한 유명한 예측을 살펴봅시다.
- 내 생각에 컴퓨터에 대한 전세계의 수요는 다섯 대 정도일 것이다. (Chairman of IBM, 1943)
- 미래에는 컴퓨터의 무게가 아마도 1.5톤을 넘지 않을 것이다. (Popular Mechanics, 1949)
- 개개인이 컴퓨터를 자기 집에 두고 싶어할 이유는 없다. (President, DEC, 1977)

이렇계 예측이란 것은 힘든 일입니다. 하지만 미래에 대해서 예측을 할 수 있다면, 더 많은 기회를 얻을 수 있습니다.

## 예측 가능 조건들
상대적으로 예측하기 쉬운 것도 있습니다. 내일 아침에 해가 뜰 시간은 정확하게 예측할 수 있습니다. 반대로, 내일 로또 당첨 번호는 정확하게 예측할 수 없습니다. 어떤 사건이나 수량의 예측가능성은 다음과 같은 몇 가지 요인에 의존합니다.
- 영향을 주는 요인을 얼마나 잘 이해할 수 있는지
- 사용할 수 있는 데이터가 얼마나 많은지
- 예측이 우리가 예측하려는 것에 영향을 줄 수 있는지 여부

예)
1. 전기 수요 예측의 경우는 보통 세 가지 조건이 모두 만족되기 때문에 상당히 정확할 수 있습니다. 우리는 전기 수요가 기온에 영향을 크게 받고, 휴일이나 경제적인 요인에 의해 영향을 적게 받는다는 사실을 잘 알고 있습니다. 전기 수요와 날씨 상황에 대해 시간에 따른 충분한 데이터가 있고, 전기 수요와 핵심 변수를 관련지어 잘 설명할 수 있는 모델을 세울 수 있는 기술도 있어서, 놀라울 정도로 정확하게 예측할 수 있습니다.

2. 반대로, 환율 예측의 경우에는 사용할 수 있는 데이터가 충분히 많다는 한 가지 조건만 만족됩니다. 하지만, 환율에 영향을 주는 요인을 잘 모르고, 환율 예측은 환율 자체에 직접적으로 영향을 주기도 합니다. 만약에, 환율이 증가할 것이라는 예측이 아주 잘 공개된다면, 사람들은 즉시 가격을 조정할 것입니다. 즉, 사람들이 돈을 기꺼이 지불하는 상황이 벌어져, 예측 자체 때문에 예측이 맞는 상황이 되고 맙니다. 이러한 의미에서, 환율은 점점 스스로를 예측하게 되는 셈입니다. 이것은 “효율적인 시장 가설(efficient market hypothesis)”의 한 예입니다. 따라서, 환율이 내일 올라갈지 떨어질지 예측하는 것은 던진 동전이 앞 면으로 떨어질지 뒷 면으로 떨어질지 예측하는 것과 거의 비슷합니다. 두 경우 모두, 무엇을 예측하던지 50% 가까이는 맞을 것입니다. 이러한 경우에, 예측하는 사람은 스스로의 한계를 자각할 필요가 있고, 더 잘 맞추는 것이 가능하다고 주장하지 말아야 할 필요가 있습니다.


## 용어 및 개념

### 의존성 
의존성은 이전 시점에서 동일한 변수를 가진 두 관측치의 연관성을 나타냅니다.

### 정상성(Stationarity)
이 매개변수는 계열의 평균 또는 평균값을 측정합니다. 값이 주어진 기간 동안 일정하게 유지되거나 데이터 전체에 스파이크가 있거나 이러한 값이 무한대를 향하는 경향이 있는 경우, 정상성이 아닙니다.

### 차분 
차분은 시계열을 정상화하고 자동으로 발생하는 상관 관계를 제어하는 기술입니다. 하지만 모든 시계열 분석에 차이가 필요한 것은 아니며 그렇게 하면 정확하지 않은 추정이 생성될 수 있습니다.

### Curve Fitting
회귀 방법으로서의 Curve Fitting은 선형 관계가 아닌 데이터에 유용합니다. 이러한 경우 곡선 피팅에 대한 수학 방정식은 실제 영향을 미치기에는 너무 많이 떨어지는 데이터가 시스템에서 사용하고 해석할 수 있는 고유한 공식을 사용하여 곡선으로 "회귀"되도록 합니다.

# 시계열 분석
시계열 분석이란, 시간 순서대로 정렬된 데이터에서 의미 있는 요약과 통계 정보를 추출하기 위한 노력을 의미합니다. 시계열 분석은 대부분 예측 과정에 많이 사용되지만, 과거의 행동을 진단하는 과정을 포함합니다.

- 전통적으로 시계열 데이터 분석은 AR(Autoregressive), MA(Moving average), ARMA(Autoregressive Moving average), ARIMA(Autoregressive Integrated Moving average) 모델 등을 활용해 불규칙적인 시계열 데이터에 규칙성을 부여하는 방식을 활용해왔다. 
- 최근 딥러닝 방법이 발전함에 따라, 시계열 특성을 반영할 수 있는 RNN, LSTM과 같은 모델들을 시계열 예측에 활용하여 뛰어난 성능을 보여줬다.

## 통계적 시계열 분석을 해야 하는 이유 
- 많은 시계열 예측 작업에서 머신러닝과 딥러닝의 예측성능은 매우 뛰어납니다. 하지만 시계열 통계(데이터) 분석은 아직도 많은 시계열의 문제에서 중요한 위치를 차지하고 있습니다. 
- 통계적 시계열 분석에 대한 지식은 시계열 머신러닝, 딥러닝 지식과 병행될 경우 매우 효과적입니다. 그리고 통계적 시계열 분석 자체로도, 높은 신뢰성이 요구되는 예측 작업에서 유용하게 사용될 수 있습니다.
